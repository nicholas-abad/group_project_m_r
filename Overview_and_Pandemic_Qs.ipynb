## Pandemic Research - Create subset for 3 year olds and plot line graph by provider_type by year

```python
providers_3_year_olds = registered.loc[registered["age"] == '3-year-olds', ["time_period", "provider_type", "number_children"]]
```

```python
providers_3_year_olds = pd.DataFrame(registered.loc[registered["age"] == '3-year-olds', ["time_period", "provider_type","number_children"]])
```

```python
providers_3_year_olds
```

```python
# summing the number of children per year and provider_type
providers_3_year_olds = providers_3_year_olds.groupby(
    [providers_3_year_olds['time_period'].dt.year, 'provider_type']
)['number_children'].sum().reset_index()

providers_3_year_olds
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(
    data = providers_3_year_olds,
    x='time_period',
    y='number_children',
    hue='provider_type',
    marker='o'
)

plt.title('3-Year-Olds by Provider Type (2018-2024)', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Number of Children', fontsize=12)
plt.legend(title='Provider Type', fontsize=10)
```

## Pandemic Research - Plotting Number of Children (total) by Provider type by Year (all ages, all entitlements)

```python
# Filter data for years 2018 to 2024
filtered_data = registered[(registered['time_period'] >= '2018-01-01') & (registered['time_period'] <= '2024-12-31')]
```

```python
# Group by time_period and provider_type, summing the number_child column
grouped_data = (
    filtered_data.groupby(['time_period', 'provider_type'])['number_children']
    .sum()
    .reset_index()
)
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(
    data=grouped_data,
    x='time_period',
    y='number_children',
    hue='provider_type',
    marker='o'
)

plt.title('Number of Children by Provider Type (2018-2024)', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Number of Children', fontsize=12)
plt.legend(title='Provider Type', fontsize=10)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0)
```
