## Edit age: remove total, 2 year olds from age column

```python
def filter_out_age(df):
    """
    Removes rows where the column 'age' contains specific unwanted values.
    Parameters:
    df: The DataFrame to modify
    Returns:
    pandas.DataFrame: The modified DataFrame with rows removed.
    """
    unwanted_values = ['2-year-olds', 'Total']
    if 'age' in df.columns:
        return df[~df['age'].isin(unwanted_values)].reset_index(drop=True)
```

```python
sen = filter_out_age(sen)
```
