## Filtering out 2-year-olds on registered_15 hours df

```python
def filter_out_age(df):
    """
    Removes rows where the column 'age' is equal to '2-year-olds' from a dataframe.

    Parameters:
    df: The DataFrame to modify

    Returns:
    pandas.DataFrame: The modified DataFrame with rows removed. 
    """
    if 'age' in df.columns:
        return df[df['age']!= '2-year-olds'].reset_index(drop=True)
```

```python
# Check the count of '2-year-olds' before filtering
print(f"Before filtering, count of '2-year-olds': {(registered_15_hours['age'] == '2-year-olds').sum()}")

# Apply the filter and reassign the DataFrame
registered_15_hours = filter_out_age(registered_15_hours)

# Check the count of '2-year-olds' after filtering
print(f"After filtering, count of '2-year-olds': {(registered_15_hours['age'] == '2-year-olds').sum()}")
```

```python
def filter_out_age(df):
    """
    Removes rows where the column 'age' is equal to '2-year-olds' from a dataframe.

    Parameters:
    df: The DataFrame to modify

    Returns:
    pandas.DataFrame: The modified DataFrame with rows removed. 
    """
    if 'age' in df.columns:
        return df[df['age']!= '3 and 4-year-olds'].reset_index(drop=True)
```

```python
# Check the count of '3 and 4-year-olds' before filtering
print(f"Before filtering, count of '3 and 4-year-olds': {(registered_15_hours['age'] == '3 and 4-year-olds').sum()}")

# Apply the filter and reassign the DataFrame
registered_15_hours = filter_out_age(registered_15_hours)

# Check the count of '3 and 4-year-olds' after filtering
print(f"After filtering, count of '3 and 4-year-olds': {(registered_15_hours['age'] == '3 and 4-year-olds').sum()}")
```

```python
registered_15_hours_columns_check = ['age']
#disadvantaged_columns_check = ['age',, 'disadvantage_type', 'basis_for_funding']

def remove_rows_with_total(df, columns_to_check):
    """
    Removes rows containing the value 'total' in specified columns.

    Parameters:
    df (pandas.DataFrame): The DataFrame to process.
    columns_to_check (list): List of column names to check for the value 'total'.

    Returns:
    pandas.DataFrame: The cleaned DataFrame with rows containing 'total' removed.
    """
    indices_to_remove = []

    for idx, row in df.iterrows():
        for column in columns_to_check:
            if str(row[column]) == 'Total':
              indices_to_remove.append(idx)
              break
    df = df.drop(indices_to_remove).reset_index(drop=True)

    return df
```

```python
registered_15_hours = remove_rows_with_total(registered_15_hours, registered_15_hours_columns_check)
```
